```python
import sys
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://www.the-numbers.com/market/'

html = requests.get(url).content
soup = BeautifulSoup(html,'lxml')
```

```python
text =soup.find_all('table')[2].get_text()
print(text)
```

```python
newtxt = text.split('\n')
```

```python
repeat=newtxt.count('')

for item in range(repeat):
    newtxt.remove(char)
    
newtxt=newtxt[1:]
print(newtxt)
```

```python
#9 values per row
Year=[]
Movie=[]
Genre1=[]
Genre2=[]
Genre3=[]
Genre4=[]
TicketsSold=[]
```

```python
for index in range(len(newtxt)):
    item = newtxt[index]
    if len(item)==4:
        Year.append(item)
    if len(newtxt[index-1])==4:
        Movie.append(item)
    if len(newtxt[index-2])==4:
        Genre1.append(item)
    if len(newtxt[index-3])==4:
        Genre2.append(item)
    if len(newtxt[index-4])==4:
        Genre3.append(item)
    if len(newtxt[index-5])==4:
        Genre4.append(item)
    if len(newtxt[index-10])==4:
        TicketsSold.append(int(item.replace(',','')))
```

```python
print(TicketsSold)
```
